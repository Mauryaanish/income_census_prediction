# Data Cleaing and Feature Engineering 

```python
# Remove White Space
def white_Space(data):
    
    for col in data.columns:

        if data[col].dtypes == "object":
                data[col] = data[col].map(str.strip)
        else:
            pass
    
white_Space(df)
```

```python
for col in df.columns:
    if df[col].dtypes =="object":
        df[col] = df[col].str.replace("-" , " ")
```

```python
## Capital Loss is not Effect in data set so i drop this columns
df.drop(labels = ["Capital-loss"] , axis = 1  , inplace = True)
```

```python
# Find Unique Value Every Columns
for col in df.columns:
    print(df[col].unique())
    print("********************************************************************************************")
```

```python
df["Work-class"].mode()
```

```python
df["Work-class"] = df["Work-class"].replace("?" , "Private")
```

```python
df["Occupation"].mode()
```

```python
df["Occuoation"] = df["Occupation"].replace("?" , "Prof specialty")
```

```python
df["Native-county"].mode()
```

```python
df["Native-county"] = df["Native-county"].replace("?" , "United States")
```

```python
df["Native-county"] = df["Native-county"].replace("Outlying US(Guam USVI etc)" , "Outlying US")
```

```python
df["Native-county"] = df["Native-county"].replace("Trinadad&Tobago" , "Trinadad Tobago")
```

```python
# income maping binnay form  <=50K = 0 , >50K = 1
df["Income"].unique()
```

```python
map_income = {" <=50K" : 0 , " >50K" : 1}
df["Income"] = df["Income"].map(map_income)
```

```python
# Divide Independent and Dependent columns

X  = df.iloc[: ,:-1]
y = df["Income"]
```

```python
# Define which columns should be ordinal- encoded which should be scaled

categorical_col = X.select_dtypes(include = "object").columns
numerical_col = X.select_dtypes(exclude="object").columns
```

```python
print(categorical_col)
print(numerical_col)
```

```python
from sklearn.impute import SimpleImputer # handling a missing Value
from sklearn.preprocessing import StandardScaler # Handling Feature Scaling
from sklearn.preprocessing import OneHotEncoder # onehotencoder

#pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# numerical pipeline

num_pipeline = Pipeline(
                steps =[
                    ("imputer" , SimpleImputer(strategy = "median")),
                    ("scaler" , StandardScaler())
                ])
cat_pipeline = Pipeline(
                steps=[
                   ("imputer" ,  SimpleImputer(strategy = "most_frequent")),
                    ("one_hot_encoder" , OneHotEncoder(sparse = False)),
                    ("scaler" , StandardScaler())
                ])

preprocessor = ColumnTransformer(
[
    ("num_pipeline" , num_pipeline ,numerical_col),
    ("cat_pipeline" , cat_pipeline ,categorical_col)
])
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X ,y ,test_size = 0.30 , random_state =30)
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train))
```

```python
X_train
```

```python
X_test = pd.DataFrame(preprocessor.transform(X_test))
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter = {'penalty' : ["l1" , 'l2' , 'elasticnet'] , "C" : [1,2,3,4,5,10,15,20] , 'max_iter' : [100,200,300]}
```

```python
classifier_regressor = GridSearchCV(classifier , param_grid = parameter , scoring="accuracy" , cv = 5)
```

```python
classifier_regressor.fit(X_train , y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
## prediction 
y_pred = classifier_regressor.predict(X_test)
```

```python
## accuracy Score
from sklearn.metrics import accuracy_score , classification_report, confusion_matrix
```

```python
score = accuracy_score(y_pred , y_test)
print(score)
```

```python
confusion_matrix(y_pred , y_test)
```

```python
print(classification_report (y_test , y_pred))
```
